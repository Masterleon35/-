## 开始

```python
raw = pd.read_excel('/Users/litianhao/Desktop/processed/completed/company-node_mapping.xlsx',header=None,na_values='##',keep_default_na=False)
```

```python
standard_list = []
not_standard_list = []
with codecs.open('/Users/litianhao/Desktop/processed/行政区/行政区三级_标准.txt','r','utf8')as f:
    for word in f:
        standard_list.append(word.strip())
with codecs.open('/Users/litianhao/Desktop/processed/行政区/行政区三级_不标准.txt','r','utf8')as f1:
    for word in f1:
        not_standard_list.append(word.strip())
standard_notstandrd_list = list(zip(not_standard_list,standard_list))
```

```python
a = HanLP.newSegment().enableCustomDictionaryForcing(True)
# a = HanLP.newSegment().enablePlaceRecognize(True)
CRFLexicalAnalyzer = JClass('com.hankcs.hanlp.model.crf.CRFLexicalAnalyzer')
#正式废弃CRFSegment，请使用快10倍的CRFLexicalAnalyzer或PerceptronLexicalAnalyzer
segment=CRFLexicalAnalyzer()
for index in range(7,399):
    yuliao_raw = raw.loc[index,2]
    if not yuliao_raw:
        continue
    if ';' in yuliao_raw:
        yuliao_raw = yuliao_raw.split(';')[0]
    elif '；'in yuliao_raw:
        yuliao_raw = yuliao_raw.split('；')[0]
    ns_list_raw = [x for x in [str(x) for x in HanLP.segment(yuliao_raw)] if '/ns' in x]
    ns_list_standard = [x.replace('/ns','') for x in ns_list_raw]
    for index1 in range(len(ns_list_standard),0,-1):
        index1 = index1-1
        if ns_list_standard[index1] in standard_list :
            if index1 != 0 and ns_list_standard[0] in standard_list:
                raw.loc[index,3] = ns_list_standard[0]+'+'+ns_list_standard[index1]
            elif index1 != 0 and index1 != 1 and ns_list_standard[0] not in standard_list:
                raw.loc[index,3] = ns_list_standard[1]+'+'+ns_list_standard[index1]
            else:
                raw.loc[index,3] = ns_list_standard[index1]
            break
        elif ns_list_standard[index1] in not_standard_list:
            index_notstand = not_standard_list.index(ns_list_standard[index1])
            ns_list_standard[index1] = standard_notstandrd_list[index_notstand][1]
            if index1 != 0 and ns_list_standard[0] in standard_list:
                raw.loc[index,3] = ns_list_standard[0]+'+'+ns_list_standard[index1]
            elif index1 != 0 and index1 != 1 and ns_list_standard[0] not in standard_list:
                raw.loc[index,3] = ns_list_standard[1]+'+'+ns_list_standard[index1]
            else:
                raw.loc[index,3] = ns_list_standard[index1]
            break
            
#         elif ns_list_standard[index1] in standard_list and ns_list_standard[index1][-1] not in ['省','市','县','区','旗子']:
#             raw.loc[index,6] = ns_list_standard[index1]
            
    
```

## 高德API

```python
import pandas as pd
import requests
import json
from tqdm import tqdm
```

```python
full_address_list = list(pd.read_excel('/Users/litianhao/Desktop/processed/completed/company-node_mapping.xlsx',header=None,keep_default_na=False).loc[:,2])
raw = pd.read_excel('/Users/litianhao/Desktop/processed/completed/company-node_mapping.xlsx',header=None,keep_default_na=False)
```

```python
p_c_ad = []
for address in tqdm(full_address_list):
    address = address.strip()
    if address:
        param = {
            'key':'820921c611b0b12418ce4bba206efeea',
            'keywords':address,
            'output':'JSON'
        }
        result = requests.get('https://restapi.amap.com/v3/place/text?',params=param)
        result_dic = json.loads(result.text)
#         print(result_dic['pois'][0]['pname'])
            if result_dic['pois']:
                if result_dic['pois'][0]['adname']:
                    p_c_ad.append(result_dic['pois'][0]['adname'])
                elif result_dic['pois'][0]['cityname']:
                    p_c_ad.append(result_dic['pois'][0]['cityname'])
                elif result_dic['pois'][0]['pname']:
                    p_c_ad.append(result_dic['pois'][0]['pname'])
                else:
                    p_c_ad.append('')
            else:p_c_ad.append('')
    else:p_c_ad.append('')
```
